```python
import numpy as np
import pandas as pd
```

```python
playlists = pd.read_csv("playlists-v4.csv")
```

```python
playlists['id'].unique()
```

# Clean out unnecessary columns


```python
playlists = playlists.drop("Unnamed: 0", axis="columns")
playlists = playlists.drop("track_href", axis="columns")
playlists = playlists.drop("type", axis="columns")
playlists = playlists.drop("analysis_url", axis="columns")
playlists = playlists.drop("uri", axis="columns")
```

# Clean out unnecessary rows

```python
duplicates = playlists[playlists.duplicated()]
```

```python
duplicates
```

```python
# playlists = playlists.drop_duplicates()

# Adapted from https://stackoverflow.com/questions/37047420/how-to-drop-rows-of-pandas-dataframe-with-same-value-based-on-condition-in-diffe
playlists = playlists.drop(playlists.index[(playlists['id'].isin(playlists.loc[playlists['id'].duplicated(), 'id'].unique()))])
```

```python
playlists
```

```python
playlists.to_csv("playlists-v4-final.csv", index = False)
```
